```python
import random
l1 = []
seed = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in range(0,999):
    sa = []
    for i in range(8):
        sa.append(random.choice(seed))
    salt = ''.join(sa)
    l1.append(salt)
 
```

```python
def auto(df, column, index):
    print('print("'+column+'统计")')
    l2 = [i for i in range(0,999)]
    for i in range(0,index):
        print(l1[i]+' = '+ df +'['+ df +'.'+column+' == '+str(l2[i])+']')
    print ('sum = 0')
    print('num_list = list()')
    for i in range(0,index):
        print('print("'+str(l2[i])+' :", len('+ l1[i] +')/len('+ df +'))')
        print('num_list.append(len('+ l1[i] +')/len('+ df +'))')
        print('sum = sum + len('+ l1[i] +')')
    print('print("总计",sum/len('+df+'))')
    
    print('l2 = [i for i in range(0,'+str(index)+')]')
    print('plt.bar(range('+str(index)+'), num_list, tick_label = l2)')
    print('plt.show()')
```

```python
auto('ad_info', 'intertype', 11)
```
